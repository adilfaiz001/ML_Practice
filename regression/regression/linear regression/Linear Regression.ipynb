```python
import matplotlib.pyplot as plt # Visualization
import seaborn as sns # For easier statistical plotting
sns.set_style("whitegrid")
```

```python
pd.set_option('precision',2)
plt.figure(figsize=(15,12))
sns.heatmap(train.drop(["SalePrice","Id"],axis=1).corr(), square=True)
plt.suptitle("Pearson Correlation Heatmap")
plt.show()
```

```python
skews = train.skew()
for col in train.drop(["Id","SalePrice"], axis=1).columns.tolist():
    if (train[col].dtype == 'float64') or (train[col].dtype == 'int64') :
        sns.distplot(train[col].dropna(), kde=False, label="skew = " + str(round(skews[col],2)) )
        plt.legend()
        plt.show()
        
del skews        
```

```python
corr_with_SalePrice = train.drop(["Id"], axis=1).corr()["SalePrice"].sort_values(ascending=False)
plt.figure(figsize=(14,7))
corr_with_SalePrice.drop("SalePrice").plot.bar()
plt.show()

del corr_with_SalePrice
```

```python
plt.figure(figsize=(12,10))
interesting_cols = ["OverallCond", "GrLivArea", "GarageCars", "FullBath", "1stFlrSF", "2ndFlrSF"]
sns.pairplot(train[interesting_cols], dropna=True)
plt.show()

del interesting_cols
```

```python
sns.pairplot(train[["OverallCond", "GrLivArea","SalePrice"]])
plt.show()
```

```python
for col in train.columns.tolist():
    if train[col].dtype == 'object':
        sns.countplot(col, data=train)
        plt.xticks(rotation=55)
        plt.show()
```
