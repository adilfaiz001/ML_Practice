```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
```

```python
train = pd.read_csv('Yes_Bank_Training.csv')
```

```python
#Process Labels into binary
def process_label(text):
    if text == 'yes':
        return 1
    else:
        return 0
```

```python
data = train
cat_vars = ['job_description','marital_status','education_details','has_default','housing_status','previous_loan','phone_type','month_of_year','poutcome_of_campaign']
for var in cat_vars:
    cat_list = 'var' + '_' + var
    cat_list = pd.get_dummies(train[var], prefix=var)
    data1 = pd.concat([data,cat_list], axis=1)
    data = data1

cat_vars = ['job_description','marital_status','education_details','has_default','housing_status','previous_loan','phone_type','month_of_year','poutcome_of_campaign']
data_vars = data.columns.values.tolist()
to_keep = [i for i in data_vars if i not in cat_vars]
```

```python
Train = data[to_keep]
```

```python
Train['outcome'] = Train['outcome'].apply(lambda x : process_label(x))
```

```python
Train.head()
```

```python
Train.describe()
```

```python
from imblearn.over_sampling import SMOTE
X = Train.loc[:, Train.columns != 'outcome']
Y = Train.loc[:, Train.columns == 'outcome']

os = SMOTE(random_state=0)
columns = X.columns
```

```python
os_data_X, os_data_Y = os.fit_sample(X,Y)
os_data_X = pd.DataFrame(data=os_data_X,columns=columns)
os_data_Y = pd.DataFrame(data=os_data_Y,columns=['outcome'])
```

```python
data_final_vars = Train.columns.values.tolist()
y = ['outcome']
X = [i for i in data_final_vars if i not in y]

from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()

rfe = RFE(logreg,20)
rfe = rfe.fit(os_data_X, os_data_Y.values.ravel())
```

```python
print(rfe.support_)
print(rfe.ranking_)
```

```python
os_data_X.columns
```

```python
select_index = np.where(rfe.ranking_ == 1)[0]
select_cols = os_data_X.columns[select_index]
train_x = os_data_X[select_cols]
train_y = os_data_Y
```

```python
X = train_x
Y = train_y
```

```python
X.shape
```

```python
#baseline model
model = Sequential()
model.add(Dense(2,input_dim=20,activation='sigmoid',kernel_initializer='normal'))
model.add(Dense(1,input_dim=2,activation='sigmoid',kernel_initializer='normal'))
model.add(Dense(1,activation='softmax'))
model.compile(optimizer='sgd',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
model.summary()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
model.fit(X_train,y_train,epochs=20,validation_data=(X_val,y_val))
```
